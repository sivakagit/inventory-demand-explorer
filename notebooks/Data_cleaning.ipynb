```python
num_cols = [
    'Order Item Discount', 'Order Item Discount Rate', 'Order Item Product Price',
    'Order Item Quantity', 'Order Item Total', 'Order Profit Per Order', 'Sales', 'Product Price'
]


df[num_cols].dtypes
```

```python
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Verify the changes
df[num_cols].dtypes
```

Standardize and clean text columns

```python
#Columns like Category Name, Customer Segment, Department Name, Market, Order Region, Order Status, Shipping Mode, Product Name can have inconsistencies in casing, extra spaces, or typos.
text_cols = ['Category Name','Customer Segment','Department Name','Market','Order Region','Order Status','Shipping Mode','Product Name']
for col in text_cols:
    df[col] = df[col].str.strip().str.title()  # removes spaces and standardizes capitalization
```

Handle categorical duplicates / inconsistent IDs

```python
#Some columns like Customer Id, Order Id, Product Card Id may repeat across orders (expected)
df.duplicated(subset=['Order Id', 'Order Item Id']).sum()
```

Clean address columns

```python
#Customer City, Customer State, Customer Country, Customer Zipcode, Customer Street can have inconsistent formats.
df['Customer City'] = df['Customer City'].str.title().str.strip()
df['Customer State'] = df['Customer State'].str.upper().str.strip()
df['Customer Country'] = df['Customer Country'].str.title().str.strip()
```

```python
df['Customer Zipcode'] = df['Customer Zipcode'].str.extract('(\d+)')
```

Handle date columns

```python
# Convert to datetime safely
date_cols = ['order date (DateOrders)', 'shipping date (DateOrders)']
df[date_cols].head()
```

```python

for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

# Check conversion result
df[date_cols].head()
```

```python
def outlier_summary(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col}: {len(outliers)} outliers ({len(outliers)/len(df)*100:.2f}%)")

for col in num_cols:
    outlier_summary(df, col)

```

```python
# After finishing data cleaning
df.to_csv('../data/cleaned/cleaned_data.csv', index=False)
```
